```python
#Data Visualization with Matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter plot with color mapping based on transaction value
plt.figure(figsize=(12, 6))
scatter = plt.scatter(df['timestamp'], df['value'], c=df['value'], cmap='viridis', s=100)
plt.colorbar(scatter, label='Transaction Value (Ether)')
plt.title('Ethereum Transactions Over a Week')
plt.xlabel('Timestamp')
plt.ylabel('Transaction Value (Ether)')
plt.xticks(rotation=45)
plt.savefig('scatter_plot.png')
plt.show()
```

```python
# Bubble chart with block number as size of bubbles
plt.figure(figsize=(12, 6))
bubble = plt.scatter(df['timestamp'], df['value'], s=df['block_number'] / 1000, alpha=0.5, c='blue')
plt.title('Ethereum Transactions Bubble Chart')
plt.xlabel('Timestamp')
plt.ylabel('Transaction Value (Ether)')
plt.xticks(rotation=45)
plt.savefig('bubble_chart.png')
plt.show()
```

```python
# Pair plot to show relationships
pairplot = sns.pairplot(df, vars=['block_number', 'value'], hue='value', palette='viridis')
plt.suptitle('Pair Plot of Ethereum Transactions', y=1.02)
pairplot.savefig('pair_plot.png')
plt.show()
```

```python
# Histogram of transaction values
plt.figure(figsize=(12, 6))
plt.hist(df['value'], bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Ethereum Transaction Values')
plt.xlabel('Transaction Value (Ether)')
plt.ylabel('Frequency')
plt.savefig('histogram.png')
plt.show()
```

```python
# Violin plot of transaction values
plt.figure(figsize=(12, 6))
sns.violinplot(x=df['value'], color='orange')
plt.title('Violin Plot of Ethereum Transaction Values')
plt.xlabel('Transaction Value (Ether)')
plt.savefig('violin_plot.png')
plt.show()
```
